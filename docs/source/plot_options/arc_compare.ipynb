Arc compare plots
=================

Arc compare plots are a flexible plot type for comparing multiple layers of RNA
data between two samples. Sequence annotations, per-nucleotide data,
inter-nucleotide data, and secondary structures can all be displayed.


```python
import rnavigate as rnav
from rnavigate.examples import rnasep_1, rnasep_2

plot = rnav.plot_arcs_compare(
    samples=[rnasep_1, rnasep_2],  # two example RNAvigate samples (SHAPE-treated RNaseP)
    sequence="ss_pdb",             # all data are sequence aligned to "ss_pdb" (for each sample)
    interactions="ringmap",        # plot interactions
)
```
